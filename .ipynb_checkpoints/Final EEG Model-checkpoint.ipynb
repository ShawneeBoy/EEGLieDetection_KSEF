```python
#Necessary imports
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

#Keras imports
from keras.models import *
from keras.layers import *
```

```python
from eegutils import utils
```

```python
#Read EEG data
PATH = "data/alldata"
x_train, y_train = utils.readDataShuffled(PATH)
```

```python
#Check shape of data
print(x_train.shape)
print(y_train.shape)
```

```python
#Find longest EEG data
max([len(i) for i in x_train])
```

```python
#Pad sequences to make them all the same length
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=1500, padding = 'post')
x_train.shape
```

```python
#Truth EEG spectrum
#(This can differ everytime this notebook is run, as the order of the data is randomized)
plt.plot(x_train[6])
plt.title("Truth EEG Spectrum")
plt.xlim(right = 800)
plt.ylim(300, 700)
plt.show()
```

```python
#Lie EEG Spectrum
#(This can differ everytime this notebook is run, as the order of the data is randomized)
plt.plot(x_train[60])
plt.title("Lie EEG Spectrum")
plt.xlim(right = 800)
plt.ylim(300, 700)
plt.show()
```

```python
#Create model
model2 = Sequential()
model2.add(LSTM(100, input_shape = (1500, 2), dropout = 0.5))
model2.add(Dense(1, activation = 'sigmoid'))
```

```python
#Model summary
model2.summary()
```

```python
#Train model
model2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model2.fit(x_train, y_train, validation_split = 0.3, epochs = 2, batch_size = 32)
```

```python
#Prediction
predictX = x_train[40].reshape((1, 1500, 2))
```

```python
model2.predict(predictX)
```

```python
#Actual value
y_train[40]
```

```python
#Plot the lie spectrum which it got correct
plt.plot(x_train[40])
plt.title("Lie EEG Spectrum: Correct")
plt.xlim(right = 800)
plt.ylim(300, 700)
plt.show()
```
